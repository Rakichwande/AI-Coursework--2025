```python
#How many rows and columns are contained in the dataset?
rows, cols = df.shape
print("The dataset contains", rows, "rows and", cols, "columns.")
```

```python
#What datatypes are contained in the dataset?

print(df.dtypes)
```

```python
#Is the dataset complete? i.e no missing values?

missing_values = df.isnull().sum()

print("Missing values per column:\n", missing_values)

if missing_values.sum() == 0:
    print("\n✅ The dataset is complete — no missing values.")
else:
    print("\n⚠️ The dataset has missing values.")
```

```python
#List the number of Movies and TV Shows among the last 20 rows of the dataset.

last_20 = df.tail(20)

type_counts = last_20['type'].value_counts()

print("Number of Movies and TV Shows among the last 20 rows:")
print(type_counts)
```
